```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_pickle('odds_data.pickle')
```

```python
df1.drop(['Rot', '1st', '2nd', '3rd', '4th', 'Open', 'ML', '2H'], axis=1, inplace=True)
```

```python
# adding games numbers
colNew = []
i = 0
for index, row in df1.iterrows():
    if index == 0:
        pass
    elif index % 2 == 0:
        i += 1
    colNew.append(i)
df1['gameno'] = colNew
```

```python
df_h = df1.loc[df1['VH'] == 'H']
df_v = df1.loc[df1['VH'] == 'V']
```

```python
df_h.rename(columns={'Date': 'd1', 'VH': 'h', 'Team': 'h_team', 'Final': 'h_final', 'Close': 'h_line'}, inplace=True)
df_v.rename(columns={'Date': 'd2', 'VH': 'v', 'Team': 'v_team', 'Final': 'v_final', 'Close': 'v_line'}, inplace=True)
```

```python
df2 = df_h.merge(df_v, left_on='gameno', right_on='gameno')
```

```python
df3 = df2.drop(['gameno'], axis=1)
```

```python
df3.replace('pk', 0, inplace=True)
```

```python
lines = df3.min(axis=1)
df3['run_line'] = lines
```

```python
df = df3
```

```python
df['fav'] = np.where(df['h_line'] == df['run_line'], 'h', 'v')
```

```python
df.drop('d2', axis=1, inplace=True)
```

```python
df['final_spread_outcome'] = df['h_final'] - df['v_final']
```

```python
df['winner'] = np.where(df['final_spread_outcome'] > 0, 'h', 'v')
```

```python
df['fav_dog'] = np.where(df['final_spread_outcome'].abs() > df['run_line'], 'fav', 'dog')
```

```python
football_numbers = [2.5,3.5,6.5,7.5,9.5,10.5,13.5,14.5,19.5,20.5]
```

```python
df_fn = df[df['run_line'].isin(football_numbers)]
```

```python
football_nums_favs = [3.5,7.5,10.5,14.5]
football_nums_dogs = [2.5,6.5,9.5,13.5]
```

```python
df_fn['prediction'] = np.where(df_fn['run_line'].isin(football_nums_favs), 'dog', 'fav')
```

```python
df_fn['outcome'] = np.where(df_fn['fav_dog'] == df_fn['prediction'], 'win', 'loss')
```
